## Inheritance in python 

* With inhertance one class can derive the properties of another class 

```python
# Creating the base class 

class Vehicle:
    #creating contructor here 
    def __init__(self,mileage, cost):
        self.mileage=mileage 
        self.cost=cost
    def show_details(self):
        print("I m a Vehicle ")
        print("mileage of Vehicle is ",self.mileage)
        print("cost of Vehicle is ",self.cost)
```

```python
# Instatiating the object for base class 
v1 = Vehicle(500,1000)
v1.show_details()
```

```python
# creating the child class 

class Car (Vehicle):
    def show_car(self):
        print("i am a car")
        
    
```

```python
# instantiating the object for child class 
c1 = Car(200,12000)
```

```python
#instantaiting the object for child class
c1.show_details()
```

```python
#invoking the child class method 
c1.show_car()
```

##### Over-riding init method 

```python
class Car2(Vehicle):
    # Over-riding init method 
    
    def __init__(self,mileage,cost,tyres,hp):
        super().__init__(mileage,cost)
        self.tyres=tyres 
        self.hp=hp
    def show_car_details(self):
        print("I am the car ")
        print("Number of tyres are ",self.tyres)
        print("value of the horse power is ",self.hp)
```

```python
# invoking show_deatils () method fromt the parent class 
c2= Car2(20,20000,4,300)
```

```python
c2.show_details()
```

```python
# invoking show_car_deatails () method from child class 
c2.show_car_details()
```

##### Type of Inheritance 

1. Single Inheritance 
2. Multiple Inheritance 
3. Multi-level Inheritance 
4. Hybrid Inheritance 

. Multiple Inheritance 

* In multiple inheritance , the child inheritance from more than 1 parent class


```python
class Parent1():
    def assign_string_one(self,str1):
        self.str1=str1
    def show_string_one(self):
        return self.str1
```

```python
class Parent2():
    def assign_string_two(self,str2):
        self.str2=str2
    def show_string_two(self):
        return self.str2
```

```python
class Derived (Parent1,Parent2):
    def assign_string_three(self,str3):
        self.str3=str3
    def show_string_three(self):
        return self.str3
```

```python
# Instantiating object of child class 
d1 = Derived()
```

```python
d1.assign_string_one("one")
d1.assign_string_two("two")
d1.assign_string_three("three")
```

```python
d1.show_string_one()
```

```python
d1.show_string_two()
```

```python
d1.show_string_three()
```

##### Multi-level inheritance 

* In multi-level inheritance, we have Parent, child , grand-child relationship

```python
class Parent():
    def assign_name(self,name):
        self.name= name
    def show_name(self):
        return self.name
```

```python
class Child(Parent):
    def assign_age(self,age):
        self.age= age
    def show_age(self):
        return self.age
    
```

```python
class Grandchild(Child):
    def assign_gender(self,gender):
        self.gender=gender
    def show_gender(self):
        return self.name
    
```

```python
gc=Grandchild()
```

```python
gc.assign_name("Nuke")
gc.assign_age(23)
gc.assign_gender("Male")
```

```python
gc.show_name()
```

```python
gc.show_age()
```

```python
 gc.show_gender()
```
